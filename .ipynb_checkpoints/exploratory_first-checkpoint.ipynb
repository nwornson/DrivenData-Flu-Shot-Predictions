```python
import pandas as pd
```

```python
train = pd.read_csv('../training_set_features.csv')
train_labels = pd.read_csv('../training_set_labels.csv')
test = pd.read_csv('../test_set_features.csv')
```

```python
train.head()
```

```python
train.info()
```

```python
# any NA's
train.isnull().values.sum()
```

```python
# where are most of the NA's located
findNA = lambda x: x.isnull().values.sum()
train.apply(findNA)
```

```python
# health_insurance, employment_industry, employment_occupation missing almost half
## change NA in health_insurance to 3
train['health_insurance'] = train['health_insurance'].fillna(3)
train['employment_industry'] = train['employment_industry'].fillna('NR')
train['employment_occupation'] = train['employment_occupation'].fillna('NR')

test['health_insurance'] = test['health_insurance'].fillna(3)
test['employment_industry'] = test['employment_industry'].fillna('NR')
test['employment_occupation'] = test['employment_occupation'].fillna('NR')


test['employment_occupation']
```

```python
#  fill numeric columns (adults, children in household) NA's with 0, fill the rest with Not Reported (NR)
train['household_adults'] = train['household_adults'].fillna(0)
test['household_adults'] = test['household_adults'].fillna(0)

train['household_children'] = train['household_children'].fillna(0)
test['household_children'] = test['household_children'].fillna(0)

train = train.fillna('NR')
test = test.fillna('NR')

train.apply(findNA)
```

```python
# change discrete values to categorical, seems to be all categorical, many are likert scale (ordinal)
#  handling likert scale data.  code as numerical if there is > x categories

# x = 4?

# make all opinions numerical

# 16-20 --> opinions, keep as float64 for now 
idx = list(range(1,16))
cols = train.columns[idx]
for col in cols:
    train[col] = train[col].astype('category')
    test[col] = train[col].astype('category')

train.info()
```

```python
# remove id
train = train.drop(columns='respondent_id')
test = test.drop(columns='respondent_id',axis=1)
```

```python
# get dummies
idx = list(range(0,35))
rem = [31,32]
idx = [r for r in idx if r not in rem]

cols = train.columns

X1 = pd.get_dummies(train,columns = cols[idx])
X2 = pd.get_dummies(test,columns = cols[idx])

```

```python
X1.info()
```

```python
# labels
train_labels.head()
```

```python
# some class imbalance
train_labels['h1n1_vaccine'].value_counts()
```

```python
train_labels['seasonal_vaccine'].value_counts()
```

```python
h1n1_Y = train_labels['h1n1_vaccine']
seasonal_Y = train_labels['seasonal_vaccine']
```

```python
# split the data into training and validation sets
from sklearn.model_selection import train_test_split
X_train_h, X_test_h, y_train_h, y_test_h = train_test_split(X1, h1n1_Y, test_size=0.2)
X_train_s, X_test_s, y_train_s, y_test_s = train_test_split(X1, seasonal_Y, test_size=0.2)
```

# Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_auc_score
# increased max iterations
model_h = LogisticRegression(max_iter=300,C=.9).fit(X_train_h,y_train_h)
```

```python
model_h.n_iter_
```

```python
model_h.score(X_test_h,y_test_h)
```

```python
# ROC score
preds = model_h.predict(X_test_h)
roc_auc_score(y_test_h,preds)
```

```python
model_s = LogisticRegression(max_iter=200,C=.5).fit(X_train_s,y_train_s)
```

```python
model_s.n_iter_
```

```python
model_s.score(X_test_s,y_test_s)
```

```python
# ROC score
preds = model_s.predict(X_test_s)
roc_auc_score(y_test_s,preds)
```

# Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
RF_model_h = RandomForestClassifier(n_estimators = 100,min_samples_leaf=2, max_features=171, oob_score = True)
RF_model_h = RF_model_h.fit(X_train_h,y_train_h)
```

```python
RF_model_h.score(X_test_h,y_test_h)
```

```python
preds = RF_model_h.predict(X_test_h)
roc_auc_score(y_test_h,preds)
```
